```python
from haversine import haversine
```

```python
london_coord = (51.51469, -0.092009)
```

```python
df_merged['distance_to_london_km'] = df_merged.apply(
    lambda row: haversine((row['lat'], row['lon']), london_coord), axis = 1
)
```

```python
df_merged
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = df_merged, x = 'distance_to_london_km' , y = 'Mean')
plt.title('mean vs distance to london')
plt.xlabel('Distance from London in KM')
plt.ylabel('Mean Rent')
plt.tight_layout()
plt.show()
```

```python
X = df_merged[['distance_to_london_km']]
Y = df_merged[['Mean']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
poly = PolynomialFeatures(degree = 3)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
model = LinearRegression()
model.fit(x_train_poly, y_train)
```

```python
y_pred = model.predict(x_test_poly)
```

```python
x_range = np.linspace(X.min(), X.max(), 300)
x_range_poly = poly.transform(x_range)

y_range_pred = model.predict(x_range_poly)

plt.figure(figsize=(10, 6))
plt.scatter(X, Y, label='Actual')

plt.plot(x_range, y_range_pred, color='red', linewidth=2, label='Polynomial Regression')

plt.xlabel("Distance from London (km)")
plt.ylabel("Mean Rent (£)")
plt.title("Polynomial Regression: Rent vs. Distance from London")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
y_train_pred = model.predict(x_train_poly)
y_test_pred = model.predict(x_test_poly)
```

```python
train_score = r2_score(y_train, y_train_pred)
test_score = r2_score(y_test, y_test_pred)
```

```python
df_merged['Median'] = pd.to_numeric(df_merged['Median'], errors='coerce')
df_merged['Lower quartile'] = pd.to_numeric(df_merged['Lower quartile'], errors='coerce')
```

```python
df_merged['Affordability_Score'] = df_merged['Median'] / df_merged['Lower quartile']
```

```python
df_afford = df_merged.dropna(subset=['Affordability_Score'])
```

```python
top10 = df_afford.sort_values('Affordability_Score', ascending=False).head(10)
bottom10 = df_afford.sort_values('Affordability_Score', ascending=True).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Affordability_Score', y='Area', data=top10, palette='Reds_r')
plt.title('Top 10 Least Affordable Rental Areas')
plt.xlabel('Affordability Score (Median / Lower Quartile)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Affordability_Score', y='Area', data=bottom10, palette='Greens')
plt.title('Top 10 Most Affordable Rental Areas')
plt.xlabel('Affordability Score (Median / Lower Quartile)')
plt.tight_layout()
plt.show()
```
