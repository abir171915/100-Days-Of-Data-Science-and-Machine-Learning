```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path1 = r"C:\Users\abirf\OneDrive\Desktop\waterloo.csv"
file_path2 = r"C:\Users\abirf\OneDrive\Desktop\birmingham.csv"
```

```python
df_waterloo = pd.read_csv(file_path1)
df_birmingham = pd.read_csv(file_path2)
```

```python
df_waterloo['city'] = 'Waterloo'
df_birmingham['city'] = 'Birmingham'
```

```python
df_merged = pd.concat([df_waterloo, df_birmingham], ignore_index = True)
```

```python
df_merged['datetimeLocal'] = pd.to_datetime(df_merged['datetimeLocal'])
```

```python
df_merged
```

```python
print(df_merged['parameter'].unique())
```

```python
df_no2 = df_merged[df_merged['parameter']=='no2']
```

```python
df_no2
```

```python
sns.lineplot(data=df_no2, x='datetimeLocal', y='value', hue='location_name')
plt.title("Daily NO₂ Levels: Waterloo vs Birmingham")
plt.xlabel("Date")
plt.ylabel("NO₂ (µg/m³)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
