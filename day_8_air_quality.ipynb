```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_weather_bir = pd.read_csv(r"C:\Users\abirf\OneDrive\Desktop\Birmingham wheather.csv")
```

```python
df_weather_bir.head()
```

```python
df_airq_bir = pd.read_csv(r"C:\Users\abirf\OneDrive\Desktop\birmingham.csv")
```

```python
df_airq_bir.head()
```

```python
df_airq_bir['datetimeLocal'] = pd.to_datetime(df_airq_bir['datetimeLocal'])
df_weather_bir['datetime'] = pd.to_datetime(df_weather_bir['datetime'])
```

```python
df_airq_bir.rename(columns={'datetimeLocal': 'datetime'}, inplace = True)
```

```python
df_airq_bir['datetime'] = pd.to_datetime(df_airq_bir['datetime']).dt.tz_localize(None)
df_weather_bir['datetime'] = pd.to_datetime(df_weather_bir['datetime']).dt.tz_localize(None)
```

```python
df_merged = pd.merge(df_airq_bir, df_weather_bir, on = 'datetime', how= 'inner')
```

```python
df_merged.head(20)
```

```python
df_no2 = df_merged[df_merged['parameter']=='no2']
```

```python
df_no2.columns
```

```python
df_corr = df_no2[['datetime', 'value', 'temp', 'humidity', 'windspeed']]
```

```python
df_corr.rename(columns ={'value': 'No2'}, inplace =  True)
```

```python
df_corr
```

```python
corr_matrix = df_corr.drop(columns='datetime').corr()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Between NOâ‚‚ and Weather Factors")
plt.show()
```
