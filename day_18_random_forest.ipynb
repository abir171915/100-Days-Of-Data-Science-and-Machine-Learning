```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv(r"C:\Users\abirf\OneDrive\Desktop\car_evaluation.csv", header = None)
```

```python
print(df.head())
```

```python
df.columns = ['Buying_Price', 'Maintenance_Cost', 'Number_of_Doors', 'Number_of_Persons', 'Luggage_Boot', 'Safety', 'Acceptibility' ]
```

```python
X = df.drop('Acceptibility', axis = 1)
```

```python
y = df['Acceptibility']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 20,  random_state=42)
```

```python
# One-hot encode all features
ct = ColumnTransformer(
    transformers=[
        ("encoder", OneHotEncoder(handle_unknown='ignore'), X_train.columns)
    ], remainder='passthrough'
)
```

```python
X_Train_encoded = ct.fit_transform(X_train)
X_Test_encoded = ct.fit_transform(X_test)
```

```python
X_Train_encoded = X_Train_encoded.toarray()
X_Test_encoded = X_Test_encoded.toarray()
```

```python
le = LabelEncoder()
```

```python
y_Train_encoder = le.fit_transform(y_train)
y_Test_encoder = le.transform(y_test)
```

```python
clf = RandomForestClassifier(class_weight='balanced', random_state=42)
```

```python
clf.fit(X_Train_encoded, y_Train_encoder)
```

```python
y_pred = clf.predict(X_Test_encoded)
```

```python
acc = accuracy_score(y_Test_encoder, y_pred)
```

```python
print(f"Test Accuracy: {acc:.2f}")
```
